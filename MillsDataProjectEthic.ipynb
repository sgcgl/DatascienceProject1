```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
popularMedia= pd.read_excel('NYC_Social_Media_Usage.xlsx')
popularMedia.head(20)
result2012 = popularMedia[popularMedia['Date Sampled']=="2012-12-12"]
result2012
```

```python
popularity = result2012.groupby('Platform').agg(np.sum)
top = popularity.sort_values(by=['Likes/Followers/Visits/Downloads'],ascending=False)
```

```python
top5 =top.head(5)
top5
```

```python
top5.plot(kind='line',color='red',figsize=(20,10))
plt.ylabel("How many Likes/Followers/Visits/Downloads ")
plt.show()
```
